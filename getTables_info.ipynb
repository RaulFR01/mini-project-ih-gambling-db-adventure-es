```python
import pandas as pd
```

```python
df = pd.read_excel('dataset/SQL Test Data.xlsx', 'Account')

df['AccountNo'] = df['AccountNo'].str.strip()

sql = open('account.sql', 'w')

sql.seek(0,2)

print(f"""
    CREATE TABLE Account (
      {df.columns[0]} VARCHAR(255),
      {df.columns[1]} INT,
      {df.columns[2]} VARCHAR(255),
      {df.columns[3]} VARCHAR(255),
      {df.columns[4]} INT,
      {df.columns[5]} FLOAT,
      {df.columns[6]} VARCHAR(255),
    )
""", file=sql)

df1_list = []
for i in range(len(df)):
    temp = (tuple(df.loc[i]))
    print(temp)
    df1_list.append(temp)

print(f"""
    INSERT INTO Account VALUES ({[value for value in df1_list]})
""", file=sql)

sql.close()
```
